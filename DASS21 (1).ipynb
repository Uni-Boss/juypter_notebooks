```python
# import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# read in the data from one of the files and create a dataframe
DASS21_df = pd.read_csv("C:/Users/Mike/$ML_PATH/DASS21.csv", index_col="Client")
```

```python
%matplotlib inline
DASS21_df.plot.barh(rot=0,figsize=(20,15), width=0.85, linewidth = 15, xlim=[0.0,3.5])
# Put a legend to the right of the current axis
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.show()
```

```python
DASS21_add1_df = DASS21_df.replace(0,-1)
DASS21_add1_df.head()
```

```python
%matplotlib inline
DASS21_add1_df.plot.barh(rot=0,figsize=(20,15), width=0.9, linewidth = 15, xlim=[0.0,4.5])
# Put a legend to the right of the current axis
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.xticks(np.arange(-1, 4.5, 1)) 
plt.show()
```

```python
%matplotlib inline
DASS21_df.plot.barh(rot=0,figsize=(20,15), stacked=True)
# Put a legend to the right of the current axis
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.show()
```

```python
DASS21_df.plot.box(rot=0, vert=False, figsize=(20,15))
```

We can plot each attribute on a separate plot over time:

```python
DASS21_df.iloc[:, 0].plot(figsize=(20,15))
plt.legend(bbox_to_anchor=(1.04,1),  loc="upper left", fontsize=22)
plt.show()
```

```python
DASS21_df.iloc[:, 0].plot(figsize=(20,15))
DASS21_df.iloc[:, 1].plot(figsize=(20,15))
plt.legend(bbox_to_anchor=(1.04,1),  loc="upper left", fontsize=22)
plt.show()
```

```python
We can track the progress from session to session and overlay them:
```

```python
DASS21_df.iloc[0, :].plot(kind='bar',alpha=0.3, color=['black'])
DASS21_df.iloc[1, :].plot(kind='bar',alpha=0.3, color=['red'])
plt.legend()
plt.show()
```

```python
corr = DASS21_df.corr()
plt.figure(figsize = (16,10))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
DASS21_df.iloc[:, 0].plot.pie(figsize = (16,10), title=DASS21_df.columns[0])

plt.ylabel('')
plt.legend(bbox_to_anchor=(1.25,1),  loc="upper left", fontsize=12)
plt.show()
```

```python
DASS21_add_1_df = DASS21_df + 1
DASS21_add_1_df.iloc[:, :1].plot.pie(figsize = (15,8), subplots=True, legend=False,title=DASS21_df.columns[0])
#plt.legend(bbox_to_anchor=(1.25,5),  loc="upper left", fontsize=12)
plt.ylabel('')
plt.xlabel('')
DASS21_add_1_df.iloc[:, 1:2].plot.pie(figsize = (15,8), subplots=True, legend=False, title=DASS21_df.columns[1])
plt.ylabel('')
plt.xlabel('')
plt.show()
```
